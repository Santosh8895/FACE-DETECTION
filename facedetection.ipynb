```python
import cv2
import os

# ==========================
# CHECK XML FILES
# ==========================

print("Current Folder:")
print(os.getcwd())

print("\nChecking XML Files:")

face_file = "haarcascade_frontalface_default.xml"
eye_file = "haarcascade_eye_tree_eyeglasses.xml"
nose_file = "haarcascade_mcs_nose.xml"

print("Face:", os.path.exists(face_file))
print("Eye :", os.path.exists(eye_file))
print("Nose:", os.path.exists(nose_file))

if not (os.path.exists(face_file) and
        os.path.exists(eye_file) and
        os.path.exists(nose_file)):
    print("\nOne or more XML files are missing!")
    raise SystemExit

# ==========================
# LOAD CASCADES
# ==========================

face_cascade = cv2.CascadeClassifier(face_file)
eye_cascade = cv2.CascadeClassifier(eye_file)
nose_cascade = cv2.CascadeClassifier(nose_file)

print("\nCascade Status:")
print("Face Loaded :", not face_cascade.empty())
print("Eye Loaded  :", not eye_cascade.empty())
print("Nose Loaded :", not nose_cascade.empty())

# ==========================
# START WEBCAM
# ==========================

cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()

    if not ret:
        print("Cannot access webcam")
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.2,
        minNeighbors=5,
        minSize=(80, 80)
    )

    for (x, y, w, h) in faces:

        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        cv2.putText(frame, "Face", (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7,
                    (255, 0, 0), 2)

        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]

        # Eye Detection
        eyes = eye_cascade.detectMultiScale(
            roi_gray,
            scaleFactor=1.05,
            minNeighbors=4,
            minSize=(25, 25)
        )

        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(
                roi_color,
                (ex, ey),
                (ex+ew, ey+eh),
                (0, 255, 0),
                2
            )
            cv2.putText(
                roi_color,
                "Eye",
                (ex, ey-5),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                (0, 255, 0),
                1
            )

        # Nose Detection
        noses = nose_cascade.detectMultiScale(
            roi_gray,
            scaleFactor=1.1,
            minNeighbors=5,
            minSize=(30, 30)
        )

        for (nx, ny, nw, nh) in noses:
            cv2.rectangle(
                roi_color,
                (nx, ny),
                (nx+nw, ny+nh),
                (0, 0, 255),
                2
            )
            cv2.putText(
                roi_color,
                "Nose",
                (nx, ny-5),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                (0, 0, 255),
                1
            )

    cv2.imshow("Face + Eye + Nose Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Detect face_Nose_Eye

import cv2

# Load Haar Cascades
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

eye_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_eye.xml'
)

# Download nose xml and keep in project folder
nose_cascade = cv2.CascadeClassifier(
    'haarcascade_mcs_nose.xml'
)

eye_glass_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_eye_tree_eyeglasses.xml'
)

# Start Webcam
cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect Faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=5
    )

    for (x, y, w, h) in faces:

        # Face Rectangle
        cv2.rectangle(
            frame,
            (x, y),
            (x+w, y+h),
            (255, 0, 0),
            2
        )

        face_roi_gray = gray[y:y+h, x:x+w]
        face_roi_color = frame[y:y+h, x:x+w]

        # Detect Eyes
        eyes = eye_cascade.detectMultiScale(face_roi_gray)

        for (ex, ey, ew, eh) in eyes:

            cv2.rectangle(
                face_roi_color,
                (ex, ey),
                (ex+ew, ey+eh),
                (0, 255, 0),
                2
            )

            cv2.putText(
                face_roi_color,
                "Eye",
                (ex, ey-5),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                (0,255,0),
                2
            )

        # Detect Nose
        noses = nose_cascade.detectMultiScale(
            face_roi_gray,
            scaleFactor=1.1,
            minNeighbors=5
        )

        for (nx, ny, nw, nh) in noses:

            cv2.rectangle(
                face_roi_color,
                (nx, ny),
                (nx+nw, ny+nh),
                (0, 0, 255),
                2
            )

            cv2.putText(
                face_roi_color,
                "Nose",
                (nx, ny-5),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                (0,0,255),
                2
            )

    cv2.imshow("Face Eye Nose Detection", frame)

    if cv2.waitKey(1) & 0xFF == 27:  # ESC key
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Detect face_EYe Glass Creation
import cv2

face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades +
    "haarcascade_frontalface_default.xml"
)

eye_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades +
    "haarcascade_eye_tree_eyeglasses.xml"
)

cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()

    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.2,
        minNeighbors=5
    )

    for (x, y, w, h) in faces:

        cv2.rectangle(frame, (x, y), (x+w, y+h), (255,0,0), 2)

        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]

        eyes = eye_cascade.detectMultiScale(
            roi_gray,
            scaleFactor=1.1,
            minNeighbors=5
        )

        if len(eyes) >= 2:

            eyes = sorted(eyes, key=lambda e: e[0])

            left_eye = eyes[0]
            right_eye = eyes[1]

            lx, ly, lw, lh = left_eye
            rx, ry, rw, rh = right_eye

            # Left lens
            cv2.rectangle(
                roi_color,
                (lx, ly),
                (lx+lw, ly+lh),
                (0, 255, 255),
                2
            )

            # Right lens
            cv2.rectangle(
                roi_color,
                (rx, ry),
                (rx+rw, ry+rh),
                (0, 255, 255),
                2
            )

            # Bridge between glasses
            cv2.line(
                roi_color,
                (lx+lw, ly+lh//2),
                (rx, ry+rh//2),
                (0, 255, 255),
                2
            )

            # Left temple
            cv2.line(
                roi_color,
                (lx, ly+lh//2),
                (lx-25, ly+lh//2),
                (0, 255, 255),
                2
            )

            # Right temple
            cv2.line(
                roi_color,
                (rx+rw, ry+rh//2),
                (rx+rw+25, ry+rh//2),
                (0, 255, 255),
                2
            )

    cv2.imshow("Virtual Glasses", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import imutils

# HSV range for RED color
redLower = (157, 93, 203)
redUpper = (179, 255, 255)

# Start Webcam
camera = cv2.VideoCapture(0)

while True:

    grabbed, frame = camera.read()

    if not grabbed:
        break

    # Resize frame
    frame = imutils.resize(frame, width=600)

    # Get frame dimensions
    h, w = frame.shape[:2]

    # Draw tracking zones
    cv2.line(frame, (150, 0), (150, h), (255, 0, 0), 2)
    cv2.line(frame, (450, 0), (450, h), (255, 0, 0), 2)

    cv2.putText(frame, "LEFT", (40, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1,
                (255, 0, 0), 2)

    cv2.putText(frame, "CENTER", (220, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1,
                (0, 255, 0), 2)

    cv2.putText(frame, "RIGHT", (490, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1,
                (0, 0, 255), 2)

    # Blur image
    blurred = cv2.GaussianBlur(frame, (11, 11), 0)

    # Convert to HSV
    hsv = cv2.cvtColor(blurred, cv2.COLOR_BGR2HSV)

    # Create mask
    mask = cv2.inRange(hsv, redLower, redUpper)

    # Remove noise
    mask = cv2.erode(mask, None, iterations=2)
    mask = cv2.dilate(mask, None, iterations=2)

    # Find contours
    contours = cv2.findContours(
        mask.copy(),
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
    )

    contours = contours[0] if len(contours) == 2 else contours[1]

    if len(contours) > 0:

        # Largest contour
        c = max(contours, key=cv2.contourArea)

        # Minimum enclosing circle
        ((x, y), radius) = cv2.minEnclosingCircle(c)

        # Moments
        M = cv2.moments(c)

        if M["m00"] != 0:

            center = (
                int(M["m10"] / M["m00"]),
                int(M["m01"] / M["m00"])
            )

            if radius > 10:

                # Draw circle around object
                cv2.circle(
                    frame,
                    (int(x), int(y)),
                    int(radius),
                    (0, 255, 255),
                    2
                )

                # Draw center point
                cv2.circle(
                    frame,
                    center,
                    5,
                    (0, 0, 255),
                    -1
                )

                # Display coordinates
                cv2.putText(
                    frame,
                    f"X:{center[0]} Y:{center[1]}",
                    (10, h - 20),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.7,
                    (255, 255, 255),
                    2
                )

                # Decision Logic
                if radius > 250:

                    status = "STOP"

                else:

                    if center[0] < 150:
                        status = "LEFT"

                    elif center[0] > 450:
                        status = "RIGHT"

                    else:
                        status = "FRONT"

                # Show status
                cv2.putText(
                    frame,
                    status,
                    (220, 100),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1.5,
                    (0, 255, 255),
                    3
                )

                print(
                    f"Center: {center}  Radius: {radius:.2f}  Direction: {status}"
                )

    # Show output
    cv2.imshow("Red Object Tracking", frame)
    cv2.imshow("Mask", mask)

    key = cv2.waitKey(1) & 0xFF

    if key == ord("q"):
        break

# Cleanup
camera.release()
cv2.destroyAllWindows()
```

```python
#Detect Blue Object and Capture its movement
import cv2
import imutils

# HSV Range for BLUE Color
blueLower = (100, 150, 50)
blueUpper = (140, 255, 255)

# Start Webcam
camera = cv2.VideoCapture(0)

while True:

    grabbed, frame = camera.read()

    if not grabbed:
        break

    # Resize Frame
    frame = imutils.resize(frame, width=600)

    # Frame dimensions
    h, w = frame.shape[:2]

    # Draw Tracking Zones
    cv2.line(frame, (150, 0), (150, h), (255, 0, 0), 2)
    cv2.line(frame, (450, 0), (450, h), (255, 0, 0), 2)

    cv2.putText(frame, "LEFT", (40, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8,
                (255, 0, 0), 2)

    cv2.putText(frame, "CENTER", (240, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8,
                (0, 255, 0), 2)

    cv2.putText(frame, "RIGHT", (500, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8,
                (0, 0, 255), 2)

    # Blur Image
    blurred = cv2.GaussianBlur(frame, (11, 11), 0)

    # Convert BGR to HSV
    hsv = cv2.cvtColor(blurred, cv2.COLOR_BGR2HSV)

    # Create Mask for Blue Color
    mask = cv2.inRange(hsv, blueLower, blueUpper)

    # Remove Noise
    mask = cv2.erode(mask, None, iterations=2)
    mask = cv2.dilate(mask, None, iterations=2)

    # Find Contours
    contours = cv2.findContours(
        mask.copy(),
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
    )

    contours = contours[0] if len(contours) == 2 else contours[1]

    if len(contours) > 0:

        # Largest contour
        c = max(contours, key=cv2.contourArea)

        # Enclosing Circle
        ((x, y), radius) = cv2.minEnclosingCircle(c)

        # Find center
        M = cv2.moments(c)

        if M["m00"] > 0:

            center = (
                int(M["m10"] / M["m00"]),
                int(M["m01"] / M["m00"])
            )

            if radius > 10:

                # Draw Circle
                cv2.circle(
                    frame,
                    (int(x), int(y)),
                    int(radius),
                    (0, 255, 255),
                    2
                )

                # Draw Center Point
                cv2.circle(
                    frame,
                    center,
                    5,
                    (0, 0, 255),
                    -1
                )

                # Show Coordinates
                cv2.putText(
                    frame,
                    f"X={center[0]} Y={center[1]}",
                    (10, h - 20),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.7,
                    (255, 255, 255),
                    2
                )

                # Direction Logic
                if radius > 150:
                    status = "STOP"

                elif center[0] < 150:
                    status = "LEFT"

                elif center[0] > 450:
                    status = "RIGHT"

                else:
                    status = "FRONT"

                # Display Status
                cv2.putText(
                    frame,
                    status,
                    (220, 90),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1.5,
                    (0, 255, 255),
                    3
                )

                print(
                    f"Center={center}, Radius={radius:.2f}, Direction={status}"
                )

    # Show Windows
    cv2.imshow("Blue Object Tracking", frame)
    cv2.imshow("Blue Mask", mask)

    # Press Q to Quit
    key = cv2.waitKey(1) & 0xFF

    if key == ord("q"):
        break

# Cleanup
camera.release()
cv2.destroyAllWindows()
```
